```python
# Install yfinance if not already installed
# pip install yfinance pandas matplotlib

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Download Reliance Industries data
# NSE ticker symbol for Reliance: "RELIANCE.NS"
# (for BSE, use "RELIANCE.BO")
# --------------------------
ticker = "RELIANCE.NS"
reliance = yf.download(ticker, start="2010-01-01", end="2025-01-01", interval="1d")
```

```python
print(reliance.head())
print(reliance.tail())
```

```python
reliance.to_csv("reliance_ohlcv.csv")
```

```python
plt.figure(figsize=(12,6))
plt.plot(reliance.index, reliance["Close"], label="Reliance Close Price")
plt.title("Reliance Industries Stock (2010â€“2025)")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.show()
```
